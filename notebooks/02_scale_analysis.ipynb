# 02 — 量表信效度分析

SEAM (Social Exercise Anxiety Measure) 和 SATAQ-4 (Sociocultural Attitudes Towards Appearance Questionnaire) 的信度检验。

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
from src.data_loader import load_survey
from src.analysis import cronbach_alpha, descriptive
from src.preprocessing import compute_scale_score
```

## 1. 加载清洗后数据

```python
df = pd.read_csv('../data/processed/survey_clean.csv')
print(f'N = {len(df)}')
```

## 2. SEAM 信度

根据 Levinson et al. (2013) 的量表条目定义 item 列名。

```python
# TODO: 替换为实际列名
seam_items = [f'seam_{i}' for i in range(1, 14)]
seam_reverse = []  # TODO: 反向计分条目

alpha_seam = cronbach_alpha(df, seam_items)
print(f'SEAM Cronbach\'s α = {alpha_seam:.3f}')
```

## 3. SATAQ-4 信度

SATAQ-4 包含多个子量表，分别检验。

```python
# TODO: 替换为实际列名和子量表分组
sataq_sub_scales = {
    'Internalization-Thin': [f'sataq_it_{i}' for i in range(1, 6)],
    'Internalization-Muscular': [f'sataq_im_{i}' for i in range(1, 6)],
    'Pressures-Media': [f'sataq_pm_{i}' for i in range(1, 5)],
    'Pressures-Peers': [f'sataq_pp_{i}' for i in range(1, 5)],
}

for name, items in sataq_sub_scales.items():
    alpha = cronbach_alpha(df, items)
    print(f'{name}: α = {alpha:.3f}')
```

## 4. 计算量表总分

```python
df['seam_score'] = compute_scale_score(df, seam_items, seam_reverse)
for name, items in sataq_sub_scales.items():
    df[f'sataq_{name.lower().replace("-", "_")}'] = compute_scale_score(df, items)

df.to_csv('../data/processed/survey_scored.csv', index=False)
print('Scored data saved.')
```
